# How to create a basic array
This section covers np.zeros(), np.ones(), np.empty(), np.arange(), np.linspace()


```python

# create 1d array of  ones
a=np.ones(10)
print(a)
# create 2d array of ones
b=np.ones((2,3))
print(b)
```

```python
# create 1d array of  zeros
a=np.zeros(4)
print(a)
# create 2d array of zeros
b=np.zeros((4,5))
print(b)
```

```python
# create empty array
e=np.empty((5))
print(e)
# create 2d empty array
d=np.empty((3,4))
print(d)
```

```python
# create array with a range of values
a=np.arange(5)
a
```

```python
# create array with a range of values with start and end
a=np.arange(1,10,3)
a
```

```python
# reshape example
a=np.arange(6)
a.reshape(3,2)
a
```

# linspace and reshape

`np.linspace` creates evenly-spaced numbers over a specified interval.
`reshape` changes array dimensions without modifying data.

```python
# linspace examples
# Create 5 evenly spaced values from 0 to 1
arr1 = np.linspace(0, 1, 5)
print("5 values from 0 to 1:")
print(arr1)

# Create 10 evenly spaced values from 0 to 10
arr2 = np.linspace(0, 10, 10)
print("\n10 values from 0 to 10:")
print(arr2)

# linspace with retstep to get step size
arr3, step = np.linspace(0, 1, 5, retstep=True)
print(f"\n5 values from 0 to 1 with step size {step}:")
print(arr3)
```
