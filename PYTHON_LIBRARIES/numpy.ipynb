# How to create a basic array
This section covers np.zeros(), np.ones(), np.empty(), np.arange(), np.linspace()


```python

# create 1d array of  ones
a=np.ones(10)
print(a)
# create 2d array of ones
b=np.ones((2,3))
print(b)
```

```python
# create 1d array of  zeros
a=np.zeros(4)
print(a)
# create 2d array of zeros
b=np.zeros((4,5))
print(b)
```

```python
# create empty array
e=np.empty((5))
print(e)
# create 2d empty array
d=np.empty((3,4))
print(d)
```

```python
# create array with a range of values
a=np.arange(5)
a
```

```python
# create array with a range of values with start and end
a=np.arange(1,10,3)
a
```
